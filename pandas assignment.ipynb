```python
#Q3. Find all rows where close price is less than 200 and low price is less than 150

df.loc[(df['close']<200) & (df['open']<150),:]
```

```python
# Q4. Create a new column name “diff” which given difference between high and low columns

df['diff']=df['high']-df['low']
df.head()
```

```python
#Q5. Select all rows where diff  > 100

df.loc[df['diff']>100,:]
```

```python
# Q6. Select all rows where open price is > mean of open column

# df['open'].mean()
df.loc[(df['open'])>df['open'].mean()]
```
